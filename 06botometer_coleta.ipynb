Cortando em listas de 50 em 50 devido ao limite da API

```python
# Yield successive n-sized
# chunks from l.
def divide_chunks(l, n):
     
    # looping till length l
    for i in range(0, len(l), n):
        yield l[i:i + n]
 
# How many elements each
# list should have
n = 100
data = list(divide_chunks(data, n))
```
